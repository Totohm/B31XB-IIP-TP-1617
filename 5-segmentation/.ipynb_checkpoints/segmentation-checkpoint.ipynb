```python
%matplotlib inline
%pprint off

from __future__ import print_function, division

import numpy as np
import matplotlib.pyplot as plt

from skimage import img_as_ubyte
from skimage.color import rgb2gray
from skimage.io import imread, imshow
from skimage.measure import label, regionprops
from skimage.morphology import binary_closing, binary_opening, disk
from skimage.filters import threshold_otsu
from skimage.transform import rescale
```

## Segmentation of toy example

In this exercise, you will have to automatically find how many type of coins are present on the picture `coins.jpg`.
Therefore, read the image as `ubyte` and display it. In order to speed-up the subsequent processes, divide the scale of the image by a factor of 4.

```python
from scipy import misc

Path = './images/' # Importing the image and display it
Coins_im = img_as_ubyte(imread(Path.__add__('coins.jpg')))
plt.figure()
plt.imshow(Coins_im)
plt.title('Original image')
plt.show()

Coins_im = misc.imresize(Coins_im, 25) #Resize the image

plt.figure() #Display the resized image
plt.imshow(Coins_im)
plt.title('Resize image')
plt.show()
```

Subsquently, threshold the image using a global Otsu's thresholding after converting the image into gray scale. Display the image.

```python
Coins_gray = rgb2gray(Coins_im) #Convert the image into gray scale and apply the Otsu transformation
Coins_Thr = threshold_otsu(Coins_gray)
Coins_Otsu = Coins_gray >= Coins_Thr

plt.figure() #Display the transformed image
imshow(Coins_Otsu)
plt.title('Image after the Otsu\'s thresholding' )
```

Clean the image using some morphological operations, i.e. erosion, dilation, opening, closing ...

```python
from skimage.morphology import erosion, dilation, opening, closing, white_tophat

selem = disk(2) #Create the filter for the operations

Coins_MOtsu = opening(Coins_Otsu, selem) #Apply different morphological opération to have a better image
Coins_MOtsu = dilation(Coins_MOtsu, selem)
Coins_MOtsu = closing(Coins_MOtsu, selem)
Coins_MOtsu = erosion(Coins_MOtsu, selem)
Coins_MOtsu = closing(Coins_MOtsu, selem)


plt.figure() #Display the processed image
imshow(Coins_MOtsu)
plt.title('Image after the Otsu\'s thresholding, and morphological operations' )
```

Using the functions [`label`](http://scikit-image.org/docs/stable/api/skimage.measure.html#skimage.measure.label) and [`regionprops`](http://scikit-image.org/docs/stable/api/skimage.measure.html#skimage.measure.regionprops) determine, the radius of each region and determine the number of type of coins in the image.

1. Show the image with the different labels.
1. How many regions do you find?
1. Display the radius of each region. Draw the corresponding circle on the image.

```python
from skimage.draw import circle

[Coins_label, label_num] = label(Coins_MOtsu, 8, True, True, 2) #Label all the region from the image

plt.figure()
imshow(Coins_label) # Show the labeled image
plt.title('Labeled image')

print('There is ', label_num, 'labels in this image') #Print the number of labels in the image
for region in regionprops(Coins_label): #Get the properties of all the labels of the image
    props = regionprops(Coins_label)
    
img = np.zeros(np.shape(Coins_label), dtype=np.uint8) #Create a new image to draw the circle

for k in range (0, label_num):
    print('The diameter of the ', k, 'th regions is', round(props[k].equivalent_diameter), 
          ' pixels, located at ', props[k].centroid) #Print the size and the location of the coins
    x, y = props[k].centroid #Store the coordinate of the coins
    rr, cc = circle(round(x), round(y), round((props[k].equivalent_diameter)/2)) #Draw the circles according to the parameters of the coins
    img[rr, cc] = 1

plt.figure() #Show the re-drawn image
imshow(img)
plt.title('Re-Drawn image')
```
